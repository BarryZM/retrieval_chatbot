```python
import urllib
import requests
from bs4 import BeautifulSoup


question = "如何学习英语"


class PyCrawler:
    def __init__(self,question = None):
        self.answer_url = []
        self.answer_text = []
        self.question = question
        self.search_url = None
        
    def get_search_url(self):
        '''构建问题搜索的url'''
        pass
        
    def get_answer_url(self):
        """爬取答案的url"""
        pass
        
    def get_answer_text(self):
        """爬取答案的文本"""
    
    def print_answer(self):
        """打印答案"""
        
    def search(self,question):
        """搜索问题"""
        self.answer_url = []
        self.answer_text = []
        self.search_url = None
        self.question = question
        self.get_search_url()
        self.get_answer_url()
        self.get_answer_text()
        self.print_answer()
    
#百度知道 
class ZhidaoCrawling(PyCrawler):
    def __init__(self, question = None):
        self.answer_url = []
        self.answer_text = []
        self.question = question
        self.search_url = None
        
    def get_search_url(self):
        data = {
        'ie': 'utf-8',
        'word': self.question
        }
        self.search_url="https://zhidao.baidu.com/search?" + urllib.parse.urlencode(data)
        print(self.search_url)
        
    def get_answer_url(self):
        html=requests.get(self.search_url)
        soup=BeautifulSoup(html.text, 'html.parser')
        html_dl_tag = soup.select('.dl')
        
        headers = {
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Encoding': 'gzip, deflate, compress',
            'Accept-Language': 'en-us;q=0.5,en;q=0.3',
            'Cache-Control': 'max-age=0',
            'Connection': 'keep-alive',
            'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:22.0) Gecko/20100101 Firefox/22.0'
        }

        for tag in html_dl_tag:
            href = tag.find('a').get('href')
            #print(href)
            baidu_url = requests.get(url=href, headers=headers, allow_redirects=False)
            real_url = baidu_url.headers['Location']  
            #print(real_url)
            if real_url.startswith('http'):
                self.answer_url.append(real_url)
        #print(self.answer_url)
        for url in self.answer_url:
            print(url)
        
    def get_answer_text(self):
        for url in self.answer_url:
            zhidao=requests.get(url)
            soup=BeautifulSoup(zhidao.text, 'html.parser')
            tags = soup.findAll(name="div", attrs={"class" :"best-text mb-10"})
            for tag in tags: 
                text = tag.get_text().strip("\n").encode(encoding=zhidao.encoding, errors='ignore').decode(encoding='gb18030', errors='ignore')
                text = text[4:].strip("\n")
                self.answer_text.append(text)
                #print(text)
        #print(self.answer_text)
        return self.answer_text 
    
    def print_answer(self):
        for i in range(len(self.answer_text)):
            print(str(i) + ". " + self.answer_text[i] + "\n")
        #print(self.answer_text)
        
    def search(self,question):
        self.answer_url = []
        self.answer_text = []
        self.search_url = None
        self.question = question
        self.get_search_url()
        self.get_answer_url()
        self.get_answer_text()
        self.print_answer()
        print("done")
        
        
        
#搜狗问问
class SouguoWenwen(PyCrawler):
    def __init__(self,question = None):
        self.answer_url = []
        self.answer_text = []
        self.question = question
        self.search_url = None
        
    def get_search_url(self):
        '''构建问题搜索的url'''
        data = {
        'ie': 'utf8',
        'query': self.question,
        'pid': 'sogou-brse-7da9599ceaa0dd24'
        }
        self.search_url="https://www.sogou.com/sogou?" + urllib.parse.urlencode(data)
        print(self.search_url)
        pass
        
    def get_answer_url(self):
        """爬取答案的url"""
        pass
        
    def get_answer_text(self):
        """爬取答案的文本"""
    
    def print_answer(self):
        """打印答案"""
        
    def search(self,question):
        """搜索问题"""
        self.answer_url = []
        self.answer_text = []
        self.search_url = None
        self.question = question
        self.get_search_url()
        self.get_answer_url()
        self.get_answer_text()
        self.print_answer()
        
        
        
        
        
BaiduZhidao = ZhidaoCrawling(question)

BaiduZhidao.search(question)
        
```
